```python
plt.figure(figsize=(16,6))
k = 13
cols = dt.corr().nlargest(k,'quality')['quality'].index
cm = dt[cols].corr()
sns.heatmap(cm,annot=True)
```

```python
values = dt.columns.values
number_of_columns = 13
number_of_rows = 13
number_of_rows
```

```python
plt.figure(figsize = (number_of_columns,5*number_of_rows))

for i in range(0,len(values)):
  plt.subplot(number_of_columns,number_of_rows,i+1)
  sns.set_style('whitegrid')
  sns.boxplot(dt[values[i]],color='pink',orient='v')
  plt.tight_layout()
```

```python
plt.figure(figsize = (2*number_of_columns,5*number_of_rows))

for i in range(0,len(values)):
  plt.subplot(number_of_rows,number_of_columns,i+1)
  sns.distplot(dt[values[i]],kde=True)
```
